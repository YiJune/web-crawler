```python
from selenium import webdriver#瀏覽器
from selenium.webdriver.common.by import By#選取器
from selenium.webdriver.support.wait import WebDriverWait #網路等待
from selenium.webdriver.support import expected_conditions as ec #元素狀態判斷
from selenium.webdriver.chrome.options import Options #瀏覽器設定
import time

driver = webdriver.Chrome("C:\\Users\\YiJune\\PY\\chromedriver.exe") #先下載chromedriver
driver.get("https://www.1111.com.tw/") #連結網頁
findtext=WebDriverWait(driver,10).until(ec.element_to_be_clickable((By.CSS_SELECTOR, "input[id=ks]")))
findtext.send_keys("大數據")
div = driver.find_element(By.ID,"searchjobBtn")
div.click()
# 自動滾動到底部
scroll_height = driver.execute_script("return document.documentElement.scrollHeight")
scroll_step = 1000  # 每次滾動的距離（像素）
current_scroll = 0

while current_scroll < scroll_height:
    try:
        scroll_step += 1
        driver.execute_script(f"window.scrollBy(0, {scroll_step})")
        current_scroll += scroll_step
        scroll_height = driver.execute_script("return document.documentElement.scrollHeight")
        time.sleep(1)
        WebDriverWait(driver, 1).until(ec.presence_of_element_located((By.TAG_NAME, 'body')))
        print(current_scroll)
    except :
        # 網頁無法滾動，關閉網頁
        driver.quit()

```

```python
from bs4 import BeautifulSoup as bs
from openpyxl import Workbook

# 創建 Excel 工作簿
wb = Workbook()
ws = wb.active
soup=bs(driver.page_source,"html.parser")
num=0
for job in soup.find_all("div",class_="body-wrapper"):
    num+=1
    if num ==job:
        break
    print(job.h5.text)
    print(job.h6.text)
    print(job.find("a",class_="job_item_detail_location mr-3 position-relative").text)
    print(job.find("div",class_="job_item_detail_salary ml-3 font-weight-style digit_6").text)
    print("能力:")
    print(job.find("p",class_="card-text job_item_description body_4").text)
    print("----------------------------------------------------------------------------------")
    a=job.h5.text
    b=job.h6.text
    c=job.find("a",class_="job_item_detail_location mr-3 position-relative").text
    d=job.find("div",class_="job_item_detail_salary ml-3 font-weight-style digit_6").text
    e=str("能力:")+job.find("p",class_="card-text job_item_description body_4").text
    ws.append([a,b,c,d,e])
    wb.save("1111Jobs.xlsx")
```
